# Problem Statement

    Use decision trees to prepare a model on fraud data 
    treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

    Data Description :

    Undergrad : person is under graduated or not
    Marital.Status : marital status of a person
    Taxable.Income : Taxable income is the amount of how much tax an individual owes to the 
    government 
    Work Experience : Work experience of an individual person
    Urban : Whether that person belongs to urban area or not

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('/content/Fraud_check.csv')
df.head(6)
```

```python
#Data Description

df.shape
```

```python
df.info()
```

```python
df.rename({'Marital.Status':'MS','Taxable.Income':'tax',
              'City.Population':'city_pop','Work.Experience':'Work'},inplace=True, axis=1)
df.head(6)
```

```python
df['tax']=df.tax.map(lambda x: 'Risky' if x <=30000
                        else 'Good')
df.head(6)
```

```python
le= LabelEncoder()
```

```python
df['Undergrad'] = le.fit_transform(df['Undergrad'])
df['MS'] = le.fit_transform(df['MS'])
df['Urban'] = le.fit_transform(df['Urban'])
df['tax']=le.fit_transform(df['tax'])
```

```python
df.head(6)
```

```python
df.info()
```

```python
x=df.iloc[:,[0,1,3,4,5]]
y=df['tax']
```

```python
x 
```

```python
y 
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x,y,train_size=0.75, random_state=143)
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
plt.rcParams['figure.dpi'] = 250

tree.plot_tree(dt, feature_names=x.columns, 
               class_names=['Taxable_income is Risky', 'Taxable_income is Good'],
              filled=True)
plt.show()
```

```python
dt = DecisionTreeClassifier(criterion='gini')
dt.fit(xtrain,ytrain)
ypred = dt.predict(xtest)
```

```python
print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

```python
tree.plot_tree(dt, feature_names=x.columns, 
               class_names=['Taxable_income is Risky', 'Taxable_income is Good'],
              filled=True)
plt.show()
```
